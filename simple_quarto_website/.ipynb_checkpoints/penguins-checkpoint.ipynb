```python
#|label  #fig-bill-scatter
import matplotlib.pyplot as plt

# Define the data for the polar plot
r = [i / 100.0 for i in range(200)]  # Values for r ranging from 0 to 2
theta = [2 * 3.1415 * i / 200.0 for i in range(200)]  # Values for theta

# Create a polar plot
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})

ax.plot(theta, r)
ax.set_rticks([0.5, 1, 1.5, 2])
ax.grid(True)

# Show the plot
plt.show()
```
